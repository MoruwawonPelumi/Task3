#Import panda as pd first then the csv files

```python
import pandas as pd

df1 = pd.read_csv('/content/orders.csv')
df2 = pd.read_csv('/content/order_details.csv')
df3 = pd.read_csv('/content/pizza_types.csv',encoding='latin')
df4 = pd.read_csv('/content/pizzas.csv')
```

#I will merge the orders (df1) and order_details(df2) dataframes using order_id to link their details.

```python
orders_details_df = pd.merge(df1,df2,on ='order_id')
orders_details_df
```

```python
orders_details_df.head()
```

#Next I'll merge the orders_details_df with the pizzas df4 dataframe using pizza_id to get the price of each pizza

```python
orders_pizzas_df = pd.merge(orders_details_df,df4,on ='pizza_id')
orders_pizzas_df
```

#Finally,I will merge orders_pizzas_df with df3 to create df

```python
df = pd.merge(orders_pizzas_df,df3,on  = 'pizza_type_id')
df
```

```python
df.head()
```

#To calculate the total revenue and total quantity of pizzas sold, I will first  calculate the revenue for each order item by multiplying quantity and  price, then sum the quantity for sales and then I will print both results.

```python
df['revenue'] = df['quantity']*df['price']

total_revenue = df['revenue'].sum()
total_revenue
```

```python
print(f"Total revenue: ${total_revenue:,.2f}")
```

##Calculate the Total quantity of Pizza sold

```python
Total_sales = df['quantity'].sum()

Total_sales
```

```python
print(f'Total quantity of pizza sold: {Total_sales}')
```

#To calculate the average price,I will identify the number of unique pizza types and then compute the average price from the df.

```python
Total_unique_orders = df['order_id'].nunique()

print(f'Total_unique_orders: {Total_unique_orders}')
```

# To identify peak sales hours I need to extract the hour from the time column in the df and create a new colum for it



```python
df['time'] = pd.to_datetime(df['time'])
df['hour'] = df['time'].dt.hour

df.head()
```

# Due to the User warning in the previous code, I will specify the format when converting the time column to datetime object.

```python
df['time'] = pd.to_datetime(df['time'], format= '%H:%M:%Sq')
df['hour'] = df['time'].dt.hour

df.head()
```

# To identify peak sales hours, I will group the df by the newly created 'hour'column and sum the revenue for each hour. Then I will plot the hourly revenue using a bar chart to visualize the peak

```python
import matplotlib.pyplot as plt

hourly_sales = df.groupby('hour')['revenue'].sum().reset_index()
plt.figure(figsize=(12,6))
plt.bar(hourly_sales['hour'], hourly_sales['revenue'], color= 'skyblue')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Revenue($)')
plt.title('Total Revenue by Hour of the Day')
plt.xticks(range(hourly_sales['hour'].min(), hourly_sales['hour'].max() + 1))
plt.grid(axis='y', linestyle = '--', alpha= 0.7 )
plt.show()
```

#To analyze the daily sales performance, I will extract the day of the week from the date column and create a new column for it and then state the date format also

```python
df['date'] = pd.to_datetime(df['date'], format = '%Y-%m-%d')
df['day_of_week']= df['date'].dt.day_name()
df.head()
```

# To complete the daily sales performance analysis. I will group the 'df' by the 'day_of_week' column, calculate total revenue for each day, and then visualize these daily sales using a bar chart.

```python
daily_sales = df. groupby('day_of_week')['revenue'].sum().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']).reset_index()


plt.figure(figsize=(10,6))
plt.bar(daily_sales['day_of_week'], daily_sales['revenue'], color ='lightcoral')
plt.xlabel('Day of the Week')
plt.ylabel('Total Revenue ($)')
plt.title('Total Revenue by Day of the Week')
plt.grid(axis='y', linestyle = '--', alpha = 0.7)
plt.show()

most_sales_day = daily_sales.loc[daily_sales['revenue'].idxmax()]
print(f'The Day with the most sales is {most_sales_day['day_of_week']} with revenue of ${most_sales_day['revenue']:,.2f}')
```

# To determine the top 5 bestselling pizzas, I will group 'df' by 'pizza_type_id', sum the quantity for each pizza,sort then in descending order and then select the top 5. After that ,I will create a bar chart to visualize the top 5 best selling pizzas

```python
top_5_pizzas = df.groupby('pizza_type_id')['quantity'].sum().nlargest(5).reset_index()

plt.figure(figsize=(10,6))
plt.bar(top_5_pizzas['pizza_type_id'],top_5_pizzas['quantity'],color='lightgreen')
plt.xlabel('Pizza Type ID')
plt.ylabel('Total Quantity Sold')
plt.title('Top 5 Bestselling Pizzas by Quantity')
plt.xticks(rotation=45 ,ha='right')
plt.grid(axis='y', linestyle='--', alpha =0.7)
plt.tight_layout()
plt.show()

print('Top 5 Bestselling Pizzas:')
print(top_5_pizzas)
```

#To analyze monthly sales trend I will extract the month from  the date column in df,calculate the total revenue for each month and then visualize these monthly sales using a line plot to identify any noticeable trends.

```python
df['month'] = df['date'].dt.month_name()

monthly_sales = df.groupby('month')['revenue'].sum().reindex(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']).reset_index()

plt.figure(figsize=(12,6))
plt.plot(monthly_sales['month'], monthly_sales['revenue'],marker= 'o', color='blue')
plt.xlabel('Month')
plt.ylabel('Total Revenue ($)')
plt.title('Total Revenue by Month')
plt.xticks(rotation=45, ha='right')
plt.grid(axis ='y', linestyle='--', alpha=0.7)
plt. tight_layout()
plt.show()

print('Monthly Sales Trends:')
print(monthly_sales)
```

# To identify underperforming types, I will group the 'df' by pizza_type_id,sum the quantity for each pizza type,sort them in descending order,and them select the bottom 5. After which I will create a bar chart to visualize these underperforming pizza types.

```python
bottom_5_pizzas = df.groupby('pizza_type_id')['quantity'].sum().nsmallest(5).reset_index()

plt.figure(figsize=(10,6))
plt.bar(bottom_5_pizzas['pizza_type_id'], bottom_5_pizzas['quantity'], color='salmon')
plt.xlabel('Pizza Type ID')
plt.ylabel('Total Quantity Sold')
plt.title('Top 5 Underperforming Pizzas by Quantity')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print('Top 5 Underperforming Pizzas:')
print(bottom_5_pizzas)
```
